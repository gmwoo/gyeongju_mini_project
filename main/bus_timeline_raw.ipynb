```python
def time_convert(time):
    hour = 0
    minute = 0
    second = 0
    if time >= 60:
        hour = time // 60
        minute = time % 60
    else:
        minute = time
    return [hour, minute]
```

```python
df['실제도착1'] = df['도착시간']
df['실제도착2'] = df['도착시간']
```

```python
import datetime
for n in range(1, 50):
    a = time_convert(df['도착시간'][n])
    #print(a[1])
    df['실제도착1'][n] = str(int(df['도착시간'][0].split(':')[0]) + a[0]) + ':' + str(int(df['도착시간'][0].split(':')[1]) + a[1])
    df['실제도착2'][n] = datetime.datetime.strptime(df['실제도착1'][n], '%H:%M').time()

df
```
